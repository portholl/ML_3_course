```python
f.publishedAt = pd.to_datetime(f.publishedAt)
```

```python
f.publishedAt.dt.date
```

```python
import matplotlib.pyplot as plt #     

# your code here
f_i = f.drop_duplicates(subset='video_id')
date = f_i.groupby(f['publishedAt'].dt.date)['video_id'].count()

plt.figure(figsize=(10,5))
date.plot(kind='line',color='green', marker='o', linestyle='-')
plt.xticks(fontsize=8,rotation=45)
plt.title("      ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.grid(True)
plt.show()
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
# your code here
date = f["country"].sort_values().value_counts()

plt.figure(figsize=(10,5))
date.plot.bar()
plt.xticks(fontsize=8,rotation=0)
plt.title("      ")
plt.xlabel("")
plt.ylabel(" ")
plt.show()
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

  `(scatter plot)`  -     -      (country='FR').          (comments_disabled=True).

    ?

```python
# your code here
f_i = f[f['country'] == 'FR']
f_i = f_i[f_i['comments_disabled'] == False]
plt.figure(figsize=(10,5))
plt.scatter(f_i['view_count'],f_i['comment_count'], s = 10)
plt.xticks(fontsize=8,rotation=0)
plt.xscale('log')
plt.yscale('log')
plt.title("       ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.show()
```
